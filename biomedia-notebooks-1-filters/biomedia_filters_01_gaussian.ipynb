```python
import math
import numpy as np
import matplotlib.pyplot as plt


from skimage import io
from skimage import filters
```

## Apply a Gaussian smoothing filter to an image

Generate a Gaussian kernel

```python
A = np.zeros((12, 12))
A[1, 1] = 1
kernel = filters.gaussian(A, sigma=0.4)
plt.imshow(kernel)
```

<div class="alert alert-block alert-success"><b>Explore:</b> Try to generate a filter for a much larger value of sigma. Adjust the size of the kernel accordingly. Explore what happens when the filter is applied to pixels at the boundary of the image.</div> 

```python
image = io.imread("../images/1200px-X-ray_of_normal_hand_by_dorsoplantar_projection.jpg")
```

```python
filtered_image = filters.gaussian(image, sigma=8)
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(filtered_image, cmap='gray')
plt.show()
```

<div class="alert alert-block alert-success"><b>Explore:</b> Compare the two images and observe how the Gaussian smoothing removes some of the noise in the image.</div>
